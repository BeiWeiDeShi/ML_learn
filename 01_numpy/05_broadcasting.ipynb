## 原地修改
当 NumPy 在执行类似于a + b的运算时，加法的结果会是新的数组，这
个新数组是需要分配新的内存空间的。要是这些数组的容量都很大的话，重新
分配空间的代价会比较昂贵。

所谓原地修改，指的是将数组间运算的结果存在原数组的位置上，不再额外分配
内存空间。

```python
##keep
x = np.arange(5)
y = np.zeros_like(x)
y
```
