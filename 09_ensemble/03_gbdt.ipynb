## 自定义report函数
以DataFrame的形式来查看随机搜索的结果。

```python
##keep
def report(rscv):
    df = pd.DataFrame(rscv.cv_results_)
    params = [col for col in df.columns if col.startswith('param_')]
    cols = params + ['mean_test_score']
    df = df[cols]
    outdf = pd.DataFrame(columns = ['param_name', 'param_value', 'mean_score', 'mean_std'])
    for col in params:
        for key, group in df.groupby(col):
            outdf = outdf.append({
                'param_name': col,
                'param_value': key,
                'mean_score': group['mean_test_score'].mean(),
                'mean_std': group['mean_test_score'].std()
            }, ignore_index = True)
    return outdf
```
