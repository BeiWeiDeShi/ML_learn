## 指标计算函数

```python
##keep
def get_drawdowns(series):
    prev_high, drawdowns = 0, []
    for val in series:
        prev_high = max(prev_high, val)
        dd = (val - prev_high) / prev_high if prev_high != 0 else 0
        drawdowns.append(dd if dd < 0 else 0)
    return drawdowns
```

```python
##keep
def calc_indicator(series):
    pct = series.pct_change()
    annret = np.nanmean(pct) * 242
    annvol = np.nanstd(pct) * np.sqrt(242)
    drawdowns = get_drawdowns(series)
    maxdd = np.nanmin(drawdowns)
    calmar = annret / (-maxdd)
    print(f'annret={annret:.{3}},annvol={annvol:.{3}},maxdd={maxdd:.{3}},calmar={calmar:.{3}}')
```
