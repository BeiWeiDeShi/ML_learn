# DataFrame
DataFrame可看作是由Series组成的字典，所有Series共享相同的index。使用时类似于关系数据库中的表。

```python
##keep
import pandas as pd
import numpy as np
```

## 构造DataFrame

```python
##keep
data = {'province': ['浙江', '浙江', '浙江', '广东', '广东', '广东'],
    'year': [2007, 2008, 2009, 2007, 2008, 2009],
    'GDP': [18753.73, 21462.69, 22990.35, 31777.01, 36796.71, 39482.56]}
frame = pd.DataFrame(data)
frame
```

## 添加没有数据的列

```python
##keep
frame2 = pd.DataFrame(data, columns=['year', 'province', 'GDP', 'debt'],
    index=['one', 'two', 'three', 'four',
    'five', 'six'])
frame2
```
